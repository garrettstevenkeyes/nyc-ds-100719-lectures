## 1. Handle the non-normal values

As you can see from the data dictionary, there are a number of responses that are coded with values that could throw off our descriptive statistics.  Let's clean up by mapping the responses to a more appropriate value, and handling any missing values.  

```python
#your code here

#reassign where the value of Physhlth = 0

zero_dict = {88:0,
            77: np.nan,
            99: np.nan}

df['PHYSHLTH'] = df['PHYSHLTH'].replace(zero_dict)
```

## 4. Turn the number of sick days into a classification of of chronically sick people.  

Chronically sick people account for a significant proportion of the costs for health insurers. If we can identify what leads people to be chronically sick, we can address them with specific treatments.



```python
# code to create new colum for chronically sick
#chronically sick people can be defined as people in the 80th percentile and higher

df['chron_sick'] = df['PHYSHLTH'] >= 10
chron_sick = len(df[df['chron_sick'] == True])
tot_pop = len(df['PHYSHLTH'])
p_sick = round((chron_sick/tot_pop) * 100, 2) 

print(f'{p_sick} % of people are chronically sick!')
```

## 5. Create a graphic that compares the percentage of people labeled as 'sick' in each state. 

```python
# code to calculate averages per state
state_df = df['_STATE'].reset_index()
sick_df = df['chron_sick'].reset_index()
sick_df
state_df
sick_by_state = sick_df.merge(state_df, on = 'index', how = 'right')
sick_by_state
```

```python
# create graphic to compare chronically sick by state

#connecticut sick 
conn_sick = len(df[(df['_STATE'] == 9) & (df['chron_sick'] == True)])   
conn_tot = len(df['_STATE'] == 9)
pct_conn_sick = round(conn_sick/conn_tot * 100, 2)

#new jersey sick
nj_sick = len(df[(df['_STATE'] == 34) & (df['chron_sick'] == True)])
nj_tot = len(df['_STATE'] == 34)
pct_nj_sick = round(nj_sick/nj_tot * 100, 2)

# #NY sick 
ny_sick = len(df[(df['_STATE'] == 36) & (df['chron_sick'] == True)])
ny_tot = len(df['_STATE'] == 36)
pct_ny_sick = round(ny_sick/ny_tot * 100, 2)

plt.bar(['Connecticut Sick', 'NJ Sick', 'NY Sick'],[pct_conn_sick,pct_nj_sick,pct_ny_sick])
plt.ylabel('Percentage Chronically Sick')
plt.title('Which state has the most chronic illness?');
```

## 6. Independent Variables (categorical)

Let's take the column `SMOKDAY2` and start to prepare it so that we can see how it is possibly correlated to `PHYSHLTH`.

1. Clean up the values. 
2. Calculate the descriptive statistics of the variable.
3. Generate a histogram for the variable.
4. Generate graphics that show how the responses to this variable might differ by state.
5. Generate a graphic that compares the physical health of the different groups of 'smokers'.

```python
#1
smoker_stats = df['SMOKDAY2'].value_counts(dropna=True)

#2 
smoker_stats.describe()

#3
pd.DataFrame(smoker_stats)
smokers = pd.DataFrame(smoker_stats)
smokers.plot(kind = 'bar')
plt.xlabel('Days smoking per week')
plt.ylabel('Number of responses')

#4
smokers_by_state = df.groupby('_STATE')['SMOKDAY2'].value_counts()
smokers_by_state.plot.bar().set_title('Smoking rate by state')
plt.xlabel('Smoking rates by state')
plt.ylabel('Number of responses')
plt.grid()

#5

phys_health_comp = df.groupby('SMOKDAY2')['chron_sick'].value_counts()
phys_health_comp.plot.bar()
plt.xlabel('Sickness rate grouped by smoking frequency')
plt.ylabel('Responses');
```
